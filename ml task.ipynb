```python

import cv2
import numpy as np
from keras.applications import MobileNet
model=MobileNet(weights='imagenet', include_top=False,input_shape=(64,64,3))
model.layers[0].input
model.layers[0].__class__.__name__
for layer in model.layers:
    layer.trainable=False
model.layers
```

```python
from keras.layers import Flatten
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
model.output
top_model=model.output
top_model=Flatten()(top_model)
top_model=Dense(1024, activation='relu')(top_model)
top_model=Dense(512, activation='relu')(top_model)
top_model=Dense(1, activation='sigmoid')(top_model)
```

```python
from keras.models import Model
```

```python
model.output
model.input
top_model
new_model=Model(inputs=model.input, outputs=top_model)
new_model.summary()
```

```python
new_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from keras_preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
training_set = train_datagen.flow_from_directory(
    'human_face/',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary')
test_set = test_datagen.flow_from_directory(
    'human_face/',
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary')
new_model.fit(
    training_set,
    #step_per_epoch=10,
    epochs=5,
    validation_steps=50)
```

```python
def face_dect(img):
    gray = cv2.cvtcolour(img,cv2.COLOR_BGR2GRAY)
    faces = face_dect.detectMultiscale(gray, 1.3 ,5)
    if faces is ():
        return None
    for(x,y,w,h) in faces:
        cropped_face = img[y:y+h, x:x+w]
    return cropped_face
cap = cv2.VideoCapture(0)
count = 0

while True:
    ret, frame = cap.read()
    if face_dect(frame) is not None:
        count = count+1
        face = cv2.resize(face_dect(frame), (200, 200))
        face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
        file_name_path = 'C:\\Users\\hp\\Desktop\\ml\\sheel' + str(count) + '.jpg'
        cv2.imwrite(file_name_path, face)
        cv2.putText(face_name_path, face)
        cv2.putText(face, str(count), (50,50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)

        cv2.imshow('Face Cropper', face)
    else:
        print("face not found")
        pass
    if cv2.waitKey(1) == 13 or count == 100:
        break
        
cap.release()
cv2.destroyAllWindows()
print("collecting samples complete")
```
